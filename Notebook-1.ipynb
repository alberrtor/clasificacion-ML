```python
# Importar las bibliotecas necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset (reemplaza la ruta con la tuya)
data = pd.read_csv('UCI_Credit_Card.csv')  # Cambia 'ruta_al_dataset.csv' por la ruta de tu archivo

# Mostrar las primeras filas del dataset para entender su estructura
data.head()
```

```python
# Verificar la informaciÃ³n del dataset
data.info()

# Obtener estadÃ­sticas descriptivas
data.describe()

# Verificar si hay valores faltantes
data.isnull().sum()
```

```python
# Visualizar la distribuciÃ³n de la variable objetivo
sns.countplot(x='default.payment.next.month', data=data)
plt.title('DistribuciÃ³n de la variable objetivo')
plt.show()
```

```python
# Visualizar la correlaciÃ³n entre variables
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de correlaciÃ³n')
plt.show()
```
